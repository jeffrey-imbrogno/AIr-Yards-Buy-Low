```python
import pandas as pd
```

```python
pd.options.display.max_rows = 999
```

```python
df18 = pd.read_excel(r'.../2018.xlsx')
df19 = pd.read_excel(r'.../2019.xlsx')
df20 = pd.read_excel(r'.../2020.xlsx')
```

```python
df18.head()
```

```python
train_df = df18[['T/GM', 'ADOT', 'Pts/G']]
test_df = df19[['T/GM', 'ADOT', 'Pts/G']]

X_train, y_train, X_test, y_test = train_df[['T/GM', 'ADOT']], train_df[['Pts/G']], test_df[['T/GM', 'ADOT']], test_df[['Pts/G']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(X_train, y_train)
train_scaled = pd.DataFrame(scaled_df)
scaled_df2 = scaler.transform(X_test)
test_scaled = pd.DataFrame(scaled_df2)
```

```python
from sklearn.svm import SVR
svr = SVR()
reg = svr.fit(train_scaled, y_train)
pred = svr.predict(test_scaled)
acc = round(svr.score(test_scaled, y_test)*100)
acc
```

```python
names = df20[['Player']]
```

```python
X_predict = df20[['T/GM', 'ADOT']]
y_actual = df20[['Pts/G']]
scaled_pred = scaler.transform(X_predict)
scaled_dfpred = pd.DataFrame(scaled_pred)
```

```python
pred_2020 = svr.predict(scaled_dfpred)
```

```python
pred_2020 = pd.DataFrame(pred_2020, columns=['Prediction'])
```

```python
df_prediction = names.merge(pred_2020, left_index=True, right_index=True).merge(y_actual, left_index=True, right_index=True)
```

```python
df_prediction['Difference'] = df_prediction.apply(lambda row: row['Prediction'] - row['Pts/G'] , axis=1)
```
